```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# интерактивный режим
# %matplotlib notebook

x = np.linspace(0, 10, 1000)

plt.plot(x, np.sin(x))

plt.show()
```

```python
# настройка габаритов окна

plt.figure(figsize=(8, 5)) #, facecolor="yellow", frameon=False)

x = np.linspace(0, 10, 1000)

plt.plot(x, np.sin(x))

plt.show()
```

```python
plt.figure(figsize=(8, 5))

x = np.linspace(0, 10, 1000)

plt.plot(x, np.sin(x), label="синусоида")
plt.plot(x, np.cos(x), label="косинусоида")

# настройка легенды
plt.legend(loc="upper center", bbox_to_anchor=(0.4, 1.02),
 labelcolor="green", frameon=True, fontsize="large")

# сетка
plt.grid(True)

plt.show()
```

- Гистограммы

```python
data = np.random.randint(10, size=50)
data
```

```python
plt.hist(data)
plt.show()
```

```python
# нормальное (гауссово) распределение
# m = 0, std = 2

data = np.random.randn(1000)

plt.hist(data, bins=100)

plt.show()
```

```python
plt.figure(figsize=(8, 6))

m = -1
std = 0.5

data_1 = np.random.normal(m, std, 1000)
data_2 = np.random.normal(0, 1, 1000)
data_3 = np.random.normal(2, 1, 1000)

plt.hist(data_1, bins=100, label="data_1")
plt.hist(data_2, bins=100, alpha=0.5, label="data_2", lw=2, ec="g")
n, b, p = plt.hist(data_3, bins=100, alpha=0.5, color="m", label="data_3")

plt.legend(loc="upper left", labelcolor=(0.3, 0.2, 0.9), frameon=False, fontsize="small")
plt.text(4.3, 36, f"mu={m}, \nsigma={std}", fontsize="large", alpha=0.7)
plt.xlim(-4, 6)
plt.ylim(0, 40)
plt.grid(True)
plt.title("Гистограмма")
plt.xlabel("ось x", size="large")
plt.ylabel("ось y", size="large")


plt.show()
```
