- Диаграммы рассеивания

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 50)
y = np.sin(x)

plt.scatter(x, y)

plt.show()
```

```python
x = np.random.randn(500)
y = np.random.randn(500)

# подбор цвета
# colors = np.random.random(500)
# colors = np.linspace(0, 1, 500)
colors = x

# подбор размера точек
sizes = np.random.random(500) * 100

plt.scatter(x, y, c=colors, cmap="plasma", s=sizes)

plt.show()
```

```python
# данные по населению и площади
p = {
    "Yakutsk": 318768,
    "Neryungri": 57009,
    "Viliusk": 11095,
    "Vladivostok": 600871,
    "Khabarovsk": 610305,
    "Blagoveshchensk": 225757
}
a = {
    "Yakutsk": 122,
    "Neryungri": 98.9,
    "Viliusk": 15,
    "Vladivostok": 331.2,
    "Khabarovsk": 383,
    "Blagoveshchensk":320.97
}

# координаты (широта - lat и долгота - lon)
lat = {
    "Yakutsk": 62,
    "Neryungri": 56,
    "Viliusk": 63,
    "Vladivostok": 43,
    "Khabarovsk": 48,
    "Blagoveshchensk":50
}
lon = {
    "Yakutsk": 129,
    "Neryungri": 124,
    "Viliusk": 121,
    "Vladivostok": 131,
    "Khabarovsk": 135,
    "Blagoveshchensk":127
}

data = pd.DataFrame(
    {"population":p, "area": a, "latitude": lat, "longitude": lon}
)
data
```

```python
# визуализация данных

# логарифмирование данных по населению
pop_log = np.log10(data["population"])

plt.scatter(data["longitude"], data["latitude"], s=data["area"], c=pop_log, marker="o")

# цветовая шкала
plt.colorbar(label="население (логарифм)")

plt.title("Дальневосточные города")
plt.xlabel("широта")
plt.ylabel("долгота")
plt.grid(True)

plt.show()
```

```python
x = np.random.randn(500)
y = np.random.randn(500)

# подбор цвета
# colors = np.random.random(500)
# colors = np.linspace(0, 1, 500)
colors = x

# подбор размера точек
sizes = np.random.random(500) * 100

symbol = 100

plt.scatter(x, y, c=colors, cmap="plasma", s=sizes, marker=f"${symbol}$")

plt.show()
```

```python
# объединение в одном окне графика и диаграммы рассеивания

x = np.linspace(0, 10, 1000)
y = np.sin(x)

# модуляция
m = y + np.random.randn(1000) * 0.4
z = y + m

# диаграмма рассеивания
# plt.scatter(x, z, c=np.linspace(z.min(), z.max(), 1000), cmap="PiYG", alpha=0.5)
plt.scatter(x, z, c=m, cmap="Spectral", alpha=1.0)

# график
plt.plot(x, y, c="k")

plt.show()
```
